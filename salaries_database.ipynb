# MAX BasePay and OverTimePay in 2011

```python
pd.read_sql('''SELECT MAX(BasePay), MAX(OvertimePay)
               FROM Salaries WHERE Year=2011
               ORDER BY BasePay,OvertimePay DESC;''',data)
```

```python
pd.read_sql('''SELECT Year , MAX(BasePay), MAX(OvertimePay)
               FROM Salaries
               GROUP BY Year;''',data)
```

```python
pd.read_sql('''SELECT BasePay FROM Salaries WHERE Year=2014
                ORDER BY BasePay DESC;''',data)
```

```python
pd.read_sql('''SELECT Id FROM Salaries
               WHERE BasePay='Not Provided' AND Year=2014;''',data)
```

```python
pd.read_sql('''SELECT Year , MAX(BasePay), MAX(OvertimePay)
               FROM Salaries
               WHERE Id NOT IN(148647,148651,148652,148653)
               GROUP BY Year;''',data)
```

```python
pd.read_sql('''SELECT Id FROM Salaries
               WHERE BasePay='' ''',data)
```

```python
pd.read_sql('''SELECT Year ,MAX(BasePay) Max_BasePay, MAX(OvertimePay) Max_OverTime FROM Salaries
               WHERE Id NOT IN(148647,148651,148652,148653)
               AND BasePay NOT LIKE '' GROUP BY Year;  ''',data)
```

# The Name who has highest BasePay by Year

```python
pd.read_sql('''SELECT Year, EmployeeName ,MAX(BasePay) FROM Salaries
               WHERE Id NOT IN(148647,148651,148652,148653)
               AND BasePay NOT LIKE '' GROUP BY Year;  ''',data)
```

# # The Name who has highest OverTimePay by Year

```python
pd.read_sql('''SELECT Year, EmployeeName ,MAX(OvertimePay) Max_OverTime FROM Salaries
               WHERE Id NOT IN(148647,148651,148652,148653)
               GROUP BY Year;  ''',data)
```

# Max otherPay by Year

```python
pd.read_sql('''SELECT Year , MAX(OtherPay) AS MAX_OtherPay
               FROM Salaries
               WHERE Id NOT IN(148647,148651,148652,148653)
               AND BasePay NOT LIKE ''
               GROUP BY Year;''',data)
```

# Average TotaPay

```python
pd.read_sql('''SELECT Year , AVG(TotalPay) FROM Salaries
               GROUP BY Year
               ORDER BY TotalPay DESC; ''',data)
```

# Sum of BasePay by Year

```python
pd.read_sql('''SELECT year,SUM(basepay) SUM_of_BasePay FROM Salaries GROUP BY year ;''',data)
```
